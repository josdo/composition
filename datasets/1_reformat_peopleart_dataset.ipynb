```python
# Update image paths of splits
for split in ['test', 'train', 'trainval', 'val']:
    fname = './PeopleArt/person_{}.txt'
    with open(fname.format(split+'_flat'), 'w') as writer:
        with open(fname.format(split), 'r') as reader:
            for line in reader:
                writer.write(line.split(' ')[0].replace('/', '_')+'\n')
```

```python
# Store label paths as splits
for split in ['test_flat', 'train_flat', 'val_flat']:
    fname = './PeopleArt/person_{}.txt'
    with open(fname.format(split+'labels'), 'w') as writer:
        with open(fname.format(split), 'r') as reader:
            for line in reader:
                writer.write(line.split('.jpg')[0] + '.jpg.xml\n')
```
